**Q-6** Plot the bargraph between Count of target variable per category.

```python
## Write your code here

cat_cols = [col for col in df.columns if df[col].dtype == 'object'
            and col != 'Churn'
            and col !='customerID'
            and col != 'TotalCharges']

for col in cat_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(data=df, x=col, hue='Churn')
    plt.title(f'Churn count by {col}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

**Q-14** Build a corelation of all predictors with 'Churn' and plot a bar chart (Use dataframe.corr() function)

```python
## Write your code here
#First see numerical predictor columns

df_numerical = df.select_dtypes(include=['float64', 'int64'])
```

```python
churn_corr=df_numerical.corr()
churn_corr
```

```python
plt.figure(figsize=(8,5))
churn_corr.plot(kind="bar")
plt.title("Correlations of Churn Predictors")
plt.xlabel("Predictors")
plt.ylabel("Correlation")
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend(loc="upper left", bbox_to_anchor=(1.02, 1))
plt.show()
```

**Q-16** Plot a line chart showing the distribution of tenure

```python
## Write your code here
tenure_dist = df['tenure'].value_counts().sort_index()

plt.figure(figsize=(8,5))
plt.plot(tenure_dist.index, tenure_dist.values)
plt.title('Distribution of Tenure')
plt.xlabel('Tenure')
plt.ylabel('Count')
plt.show()
```

**Q-17** Plot an area chart showing the distribution of tenure

```python
## Write your code here
tenure_dist = df['tenure'].value_counts().sort_index()

plt.figure(figsize=(8,5))
plt.fill_between(tenure_dist.index, tenure_dist.values)
plt.title('Distribution of Tenure')
plt.xlabel('Tenure')
plt.ylabel('Count')
plt.show()
```
